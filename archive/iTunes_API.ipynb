```python
import urllib2
import requests
import json
import re
from BeautifulSoup import BeautifulSoup
```

```python
baseUrl = 'https://itunes.apple.com/search?'
```

```python
parameters = {}
parameters['entity'] = 'podcast'
parameters['attribute'] = 'titleTerm'
parameters['term'] = 'serial'
parameters
```

```python
requestString = ''
for param in parameters:
    paramString = param + '=' + parameters[param]
    if len(requestString) == 0:
        requestString = paramString
    else:
        requestString = requestString + '&' + paramString
requestString
```

```python
searchUrl = baseUrl + requestString
searchUrl
```

```python
r = requests.get(searchUrl)
```

```python
data = r.json()
data = data['results']
data = data[0]
```

```python
def convert_unicode(input):
    if isinstance(input, dict):
        return {convert(key): convert(value) for key, value in input.iteritems()}
    elif isinstance(input, list):
        return [convert(element) for element in input]
    elif isinstance(input, unicode):
        return input.encode('utf-8')
    else:
        return input

data = convert_unicode(data)
```

```python
# scrape itunes page for ratings
itunesUrl = 'https://itunes.apple.com/us/podcast/id'
scrapeUrl = itunesUrl + str(data['collectionId'])
soup = BeautifulSoup(urllib2.urlopen(scrapeUrl).read())
meanRating = convert(re.findall(r'Customer Ratings(\d\.\d+)', soup.text))
numRatings = convert(re.findall(r'&nbsp;(\d+) Ratings', soup.text))
```
